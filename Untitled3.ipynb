```python
import pandas as pd

# Load the dataset
data_path = "/content/drive/MyDrive/Lab Exercise/diabetes.csv"
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
# Define input shape
input_shape = df.shape[1]  # Number of features in the dataset

# Define encoding dimension
encoding_dim = 2  # You can adjust this dimension as needed

# Define input layer
input_img = Input(shape=(input_shape,))

# Define encoding layer
encoded = Dense(encoding_dim, activation='relu')(input_img)

# Define decoding layer
decoded = Dense(input_shape, activation='sigmoid')(encoded)

# Define autoencoder model
autoencoder = Model(input_img, decoded)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Print model summary
autoencoder.summary()
```

```python
# Normalize the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# Train the autoencoder
autoencoder.fit(scaled_data, scaled_data, epochs=50, batch_size=32)
```
